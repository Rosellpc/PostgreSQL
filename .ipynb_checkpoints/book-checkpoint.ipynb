```python
import pandas as pd
import psycopg2
import matplotlib.pyplot as plt
import seaborn as sns
from psycopg2 import OperationalError

# Configuración visual de las gráficas
sns.set_theme(style="whitegrid")

conn_params = {
    "dbname": "dvdrental",
    "user": "postgres",
    "password": "holacaradebola",
    "host": "127.0.0.1",
    "port": "5432",
}

try:
    conn = psycopg2.connect(**conn_params)
    
    # 1. Consulta para obtener ganancias por fecha (agregación)
    query = """
    SELECT 
        CAST(payment_date AS DATE) as fecha,
        SUM(amount) as total_ganancia
    FROM payment
    GROUP BY CAST(payment_date AS DATE)
    ORDER BY fecha;
    """
    
    with conn:
        with conn.cursor() as cursor:
            cursor.execute(query)
            resultados = cursor.fetchall()
            columnas = [desc[0] for desc in cursor.description]

    # 2. Crear el DataFrame
    df_ganancias = pd.DataFrame(resultados, columns=columnas)
    df_ganancias['fecha'] = pd.to_datetime(df_ganancias['fecha'])

    # 3. Mostrar el total global en texto
    total_historico = df_ganancias['total_ganancia'].sum()
    print(f"--- Ganancias Totales Históricas: ${total_historico:,.2f} ---")

    # 4. Crear la gráfica
    plt.figure(figsize=(12, 6))
    plt.plot(df_ganancias['fecha'], df_ganancias['total_ganancia'], marker='o', linestyle='-', color='b')
    
    plt.title('Evolución de Ganancias Diarias - dvdrental', fontsize=15)
    plt.xlabel('Fecha', fontsize=12)
    plt.ylabel('Ganancia ($)', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

except OperationalError as e:
    print(f"Error de conexión: {e}")
finally:
    if conn:
        conn.close()
```
