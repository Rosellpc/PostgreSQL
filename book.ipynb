```python
import os
import pandas as pd
import psycopg2
from psycopg2 import OperationalError
from dotenv import load_dotenv

load_dotenv()

conn_params = {
    "dbname": os.getenv("PGDATABASE", "dvdrental"),
    "user": os.getenv("PGUSER", "postgres"),
    "password": os.getenv("PGPASSWORD", ""),
    "host": os.getenv("PGHOST", "127.0.0.1"),
    "port": os.getenv("PGPORT", "5432"),
}

try:
    conn = psycopg2.connect(**conn_params)
except OperationalError as e:
    raise RuntimeError(
        f"No se pudo conectar a PostgreSQL en {conn_params['host']}:{conn_params['port']}. "
        "Verifica que el servidor este iniciado y que el puerto/credenciales sean correctos."
    ) from e

with conn:
    with conn.cursor() as cursor:
        cursor.execute("SELECT * FROM customer order by customer_id desc limit 10;")
        resultados = cursor.fetchall()
        columnas = [desc[0] for desc in cursor.description]

df = pd.DataFrame(resultados, columns=columnas)
df
```
